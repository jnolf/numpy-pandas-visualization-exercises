```python
#1
import pandas as pd
import numpy as np

np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})
```

```python
#Rename to city
mpg = mpg.rename(columns={'cty': 'city'})
mpg
```

```python
#Rename to highway
mpg = mpg.rename(columns={'hwy': 'highway'})
mpg
```

```python
#Cars with higher city mpg vs highway mpg
mpg[mpg.city>mpg.highway]
# 0 outputs = no cars with higher city vs highway
```

```python
# New column for 'mileage_difference'
mpg['mileage_difference']= mpg.highway - mpg.city
mpg.head()
```

```python
#Car with the highet mileage difference
mpg[mpg.mileage_difference == mpg.mileage_difference.max()]
```

```python
lowest_compact_highway = mpg[mpg['class'] == 'compact'].nsmallest(n=1, columns = 'highway', keep= 'all')
lowest_compact_highway
```

```python
lowest_compact_highway = mpg[mpg['class'] == 'compact'].nlargest(1, 'highway', keep= 'all')
lowest_compact_highway
```

```python
mpg['average_mileage'] = mpg[['city', 'highway']].mean(axis=1)
mpg
```

```python
mpg.manufacturer.value_counts()
```

```python
best_dodge_average_mpg = mpg[mpg['manufacturer'] == 'dodge'].nlargest(1, 'average_mileage', keep= 'all')
best_dodge_average_mpg.head()
```

```python
worst_dodge_average_mpg = mpg[mpg['manufacturer'] == 'dodge'].nsmallest(1, 'average_mileage', keep= 'all')
worst_dodge_average_mpg.head()
```

```python
half_fast_hoppers = sum((mammals[mammals.hoppers].speed) > (mammals.speed.median()))
half_fast_hoppers
```

```python
#Percent of hoppers above the median versus all mammals on list
half_fast_hoppers_percentage = (sum((mammals[mammals.hoppers].speed) > (mammals.speed.median()))/(len(mammals)))*100
round(half_fast_hoppers_percentage, 2)
```
